```python
import numpy as np
import pandas as pd
from os.path import join
import os
```

```python
regions = ['borde_rural', 'borde_soacha', 'mixco_1_and_ebenezer', 'mixco_3', 'dennery']
```

```python

columns = ['id']
df_ids = pd.DataFrame(columns=columns)
for region in regions:
    test_dir = join('..', '..', 'data', region, 'roofs_test')
    # Initialize a dataframe for the features    
    

    # Walk through all images
    for root, dirs, files in os.walk(test_dir):
        for file in files:
            img_fp = join(root, file)
            id = file.split('.')[0]
            df_ids = df_ids.append({'id': id}, ignore_index=True)
```
